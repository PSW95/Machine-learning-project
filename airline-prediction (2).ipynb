```python
# importing library for clustering algo and finding best cluster
from sklearn.cluster import KMeans
from kneed import KneeLocator
import matplotlib.pyplot as plt

wcss = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_tr)
    wcss.append(kmeans.inertia_)

print("Length of K:", len(K))
print("Length of WCSS:", len(wcss))
kneedle = KneeLocator(K, wcss, curve='convex', direction='decreasing')
print("Optimal number of clusters (k):", kneedle.elbow)
```

```python
# plotting graph for the cluster with elbow method
plt.plot(K, wcss, marker='o')
plt.axvline(x=kneedle.elbow, color='r', linestyle='--', label=f'Optimal k = {kneedle.elbow}')
plt.title('Elbow Method with Detected Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.legend()
plt.show()
```

```python
# training on algo and adding predicted values colm as cluster inn dataset
km = KMeans(n_clusters=4, random_state=42)
km_add = km.fit_predict(scaled_tr)
scaled_tr['Cluster'] = km_add
```

```python
scaled_tr
```

```python
# plotting cluster classification graph
import seaborn as sns
plt.figure(figsize=(6,5))
sns.scatterplot(x=scaled_tr.iloc[:,0], y=scaled_tr.iloc[:,1], hue=km_add, palette='viridis')
plt.title('K-Means Clusters')
plt.show()
```

```python
# analysing cluster profiles
cluster_profile = scaled_tr.groupby('Cluster').mean()
print(cluster_profile)
```

```python
# selecting some randome sample data for the hierarchical model
sample_data = scaled_tr.sample(n=3000, random_state=42)
```

```python
from sklearn.cluster import AgglomerativeClustering
agg_model = AgglomerativeClustering(n_clusters=4, linkage='ward')
hier_labels = agg_model.fit_predict(sample_data)
```

```python
sample_data['HierCluster'] = hier_labels
```

```python
hier_profile = sample_data.groupby('HierCluster').mean()
print("\nCluster Profile (Mean Values per Cluster):")
print(hier_profile)
```

```python
# plotting graph with the pca
from sklearn.decomposition import PCA
import seaborn as sns

pca = PCA(n_components=2)
pca_result = pca.fit_transform(sample_data.drop('HierCluster', axis=1))
sample_data['PCA1'] = pca_result[:, 0]
sample_data['PCA2'] = pca_result[:, 1]

plt.figure(figsize=(8,6))
sns.scatterplot(data=sample_data, x='PCA1', y='PCA2', hue='HierCluster', palette='viridis')
plt.title('Agglomerative Clustering (PCA Projection - Sampled Data)')
plt.show()
```
